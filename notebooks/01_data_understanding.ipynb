## Check Data Distribution

```python
# Select numeric columns only
numeric_cols = df.select_dtypes(include=['float64','int64']).columns

print(numeric_cols)
```

```python
import matplotlib.pyplot as plt

# Plot histograms
df[numeric_cols].hist(figsize=(15,12), bins=30)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,12))

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(5,4,i)
    df.boxplot(column=col)

plt.tight_layout()
plt.show()
```

```python
df[numeric_cols].skew().sort_values(ascending=False)
```

## Correlation Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix
corr = df[numeric_cols].corr()

# Show top correlations for interpretation
print(corr)
```

```python
plt.figure(figsize=(15,12))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# BALANCE vs CREDIT_LIMIT
sns.scatterplot(x="CREDIT_LIMIT", y="BALANCE", data=df)
plt.title("BALANCE vs CREDIT_LIMIT")
plt.show()

# PURCHASES vs PAYMENTS
sns.scatterplot(x="PURCHASES", y="PAYMENTS", data=df)
plt.title("PURCHASES vs PAYMENTS")
plt.show()

# CASH_ADVANCE vs PRC_FULL_PAYMENT
sns.scatterplot(x="CASH_ADVANCE", y="PRC_FULL_PAYMENT", data=df)
plt.title("CASH_ADVANCE vs PRC_FULL_PAYMENT")
plt.show()
```
